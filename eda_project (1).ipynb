```python
import pandas as pd

df = pd.read_csv("movies.csv")
df['content_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
import matplotlib.pyplot as plt
plt.show()
```

A pie chart were used to analze the content types.

Insight:

most of content consist of movies


```python
average_imdb_rating_by_content_type = df.groupby('content_type')['imdb_rating'].mean().sort_values(ascending=False)
display(average_imdb_rating_by_content_type)
```

```python
import matplotlib.pyplot as plt

average_imdb_rating_by_content_type.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Average IMDb Rating by Content Type')
plt.xlabel('Content Type')
plt.ylabel('Average IMDb Rating')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

df['genre_primary'].value_counts().head(10).plot(kind='bar')
plt.title("Top Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
